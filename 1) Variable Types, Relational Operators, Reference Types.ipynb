# Range

 - Arithmetic progression of integers 
 - Positional Arguments 
    - range(start,stop,step) 
    - if one argument is passed it's assumed to be the stop, second is start, third is the step.
 X don't use range() for iterating over lists. Don't be un-pythonic!

 - Since python has strong iterative basis build into python ranges are not commonly used. Prefer to use enumerate() for counters and tuple unpacking using the .format 
 
 Using enumerate instead
  - better for counters 
  - enumerate() yields(idex,value) tuples
 
 - Often combined with Tuple unpacking

```python
"""
#1 Abusing range in the way C is used.
#2 Preferred use direct iterations over iterable objects such as lists
"""

print("#1") #1-----Abusing range in the way C is used.
s = [0,1,4,5,13]
for i in range(len(s)):
    print(s[i])

print("#2") #2----- Preferred use direct iterations over iterable objects such as lists
for v in s: 
    print(v)
```

```python
""" Enumerate:
#1 yields (index, value) tuples
#2 Tuple unpacking
"""

print("#1") #1----- yields (index, value) tuples
t = [6, 372, 8862, 14800, 2096886]
for p in enumerate(t):
    print(p)

print("#2") #2----- Tuple unpacking
for i,v in enumerate(t):
    print("i = {}, v = {}".format(i,v)) #interesting how it knows i and v ? must be positional
```

JSON can store Lists, bools, numbers, tuples and dictionaries. But to be saved into a file, all these structures must be reduced to strings. It is the string  version that can be read or written to a file. Python has a JSON module that will help converting the datastructures to JSON strings. Use the import function to import the JSON module.
> import json

The JSON module is mainly used to convert the python dictionary above into a JSON string that can be written into a file. 

> json_string = json.dumps (datastore)

The JSON module can also take a JSON string and convert it back to a dictionary structure:

> datastore - json. loads(json_string)

While the JSON module will convert strings to Python datatypes, normally the JSON functions are used to read and write directly from JSON files.

## Set

Usage Notes 
 - Unordered collection of unique, immutable objects (like the keys of a dictionary) 
 - Unique because it cannot contain duplicates, therefore a common use for a set is to remove duplicates. 
 - Each is a single object 
 - Use {} however Need to use the set constructor to initialize the an empty set. 
 - Adding a duplicate doesn’t add it nor cause an error. 
 - Literals:
    • delimited by { and }
    • single comma separated items
    • empty { } makes a dict, so for empty set use the set() constructor


```python
print("#1 Basic") #####################################

p = {6,28,496,8128,33550336}
print(p, " - is type: ", type(p),"\n")
print("Length is: ", len(p),"\n")
#NOTE: sets are immutable so can't append values p = p.append(6) 
p_extra6 = [6,6,28,496,8128,33550336]

print("#2 Empty set makes a dict, set() makes a set") #######################

empty_set = {}
print(empty_set, " - is type: ", type(empty_set))

set_constructor = set()
print(set_constructor, " - is type: ", type(set_constructor),"\n")

print("#3 Uniqueness ") #####################################

p_extra6 = [6,6,28,496,8128,33550336]
print(p_extra6, " - is type: ", type(p_extra6),"\n")
print("Length is: ", len(p_extra6),"\n")
#NOTE: sets are immutable so can't append values p = p.append(6) 
print("Call set constructor")
set_p_extra6 = set(p_extra6)
print(set_p_extra6, " - is type: ", type(set_p_extra6),"\n")
print("Length is: ", len(set_p_extra6),"\n")
```
